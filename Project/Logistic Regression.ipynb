```python
import numpy as np
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import keras
from sklearn.linear_model import LogisticRegression
```

```python
# Preparing Training Data
data = pd.read_csv(r"C:\Users\Win\Documents\Github Workspace\Machine_Learning\Project\Test & CV.csv", delimiter="\t")

X_train = data.iloc[:, :-1].values
y_train = data.iloc[:, -1].values

display(data)
# print(X_train.shape)
# print(y_train.shape)

# Preparing Test and Cross Validate Data
data = pd.read_csv(r"C:\Users\Win\Documents\Github Workspace\Machine_Learning\Project\Train Data.csv")

X_= data.iloc[:, :-1].values
y_ = data.iloc[:, -1].values

display(data)
# print(X_)
# print(y_)

X_cv, X_test, y_cv, y_test = train_test_split(X_, y_, test_size=0.5, random_state=1)


print(f"Train: {X_train.shape}")
print(f"CV: {X_cv.shape}")
print(f"Test: {X_test.shape}")
```

```python
from itertools import combinations
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

# Define feature names
gene_names = ['201506_at', '205486_at', '216638_s_at', '221619_s_at', '221672_s_at', '35148_at']

# Convert back to DataFrames if currently NumPy arrays
X_train = pd.DataFrame(X_train, columns=gene_names)
X_cv = pd.DataFrame(X_cv, columns=gene_names)
X_test = pd.DataFrame(X_test, columns=gene_names)

# Loop through all combinations
for k in range(1, len(gene_names) + 1):
    print(f"\n--- Feature Set Size: {k} ---")
    for comb in combinations(gene_names, k):
        feature_list = list(comb)

        # Select features
        X_train_sel = X_train[feature_list]
        X_cv_sel = X_cv[feature_list]
        X_test_sel = X_test[feature_list]

        # Train model
        model = LogisticRegression(max_iter=100)
        model.fit(X_train_sel, y_train)

        # Evaluate
        score_train = model.score(X_train_sel, y_train)
        score_cv = model.score(X_cv_sel, y_cv)
        score_test = model.score(X_test_sel, y_test)

        # Output results
        print(f"\nFeatures: {feature_list}")
        print(f"Train Score:   {score_cv:.4f}")
        print(f"CV Score:   {score_cv:.4f}")
        print(f"Test Score: {score_test:.4f}")
```

```python
from itertools import combinations
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# Define all 6 gene names
all_genes = ['205486_at', '221619_s_at', '221672_s_at', '216638_s_at', '35148_at', '201506_at']

# Loop over feature combinations of size 1 to 6
for k in range(1, len(all_genes) + 1):
    print(f"\n--- Feature Set Size: {k} ---")
    for comb in combinations(all_genes, k):
        feature_list = list(comb)
        
        # Extract selected features for each dataset
        X_train_sel = X_train[feature_list]
        X_cv_sel    = X_cv[feature_list]
        X_test_sel  = X_test[feature_list]
        
        # Train logistic regression
        model = RandomForestClassifier(n_estimators=100, random_state=42)
        model.fit(X_train_sel, y_train)
        
        # Evaluate
        score_cv   = model.score(X_cv_sel, y_cv)
        score_test = model.score(X_test_sel, y_test)
        
        # Print results
        print(f"\nFeatures: {feature_list}")
        print(f"CV Score:    {score_cv}")
        print(f"Test Score:  {score_test}")
```

```python
from itertools import combinations
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

# Define feature names
gene_names = ['201506_at', '205486_at', '216638_s_at', '221619_s_at', '221672_s_at', '35148_at']

# Ensure data are DataFrames with proper column names
X_train = pd.DataFrame(X_train, columns=gene_names)
X_cv = pd.DataFrame(X_cv, columns=gene_names)
X_test = pd.DataFrame(X_test, columns=gene_names)

# Loop over combinations of features
for k in range(1, len(gene_names) + 1):
    print(f"\n====== Feature Set Size: {k} ======")
    for comb in combinations(gene_names, k):
        feature_list = list(comb)

        # Select features
        X_train_sel = X_train[feature_list]
        X_cv_sel = X_cv[feature_list]
        X_test_sel = X_test[feature_list]

        # Train KNN model
        model = KNeighborsClassifier(n_neighbors=5)
        model.fit(X_train_sel, y_train)

        # Evaluate
        score_cv = model.score(X_cv_sel, y_cv)
        score_test = model.score(X_test_sel, y_test)

        # Print results
        print(f"\nFeatures: {feature_list}")
        print(f"  CV Score:   {score_cv}")
        print(f"  Test Score: {score_test}")
        print(f"  Model: KNN (k=5)")
```

```python
from itertools import combinations
from sklearn.svm import SVC
import pandas as pd

# Define feature names
gene_names = ['201506_at', '205486_at', '216638_s_at', '221619_s_at', '221672_s_at', '35148_at']

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_cv = scaler.transform(X_cv)
X_test = scaler.transform(X_test)


# Ensure the data are DataFrames with correct column names
X_train = pd.DataFrame(X_train, columns=gene_names)
X_cv = pd.DataFrame(X_cv, columns=gene_names)
X_test = pd.DataFrame(X_test, columns=gene_names)

# Loop through all combinations of features
for k in range(1, len(gene_names) + 1):
    print(f"\n====== Feature Set Size: {k} ======")
    for comb in combinations(gene_names, k):
        feature_list = list(comb)

        # Select features
        X_train_sel = X_train[feature_list]
        X_cv_sel = X_cv[feature_list]
        X_test_sel = X_test[feature_list]

        # Train SVM model (default kernel='rbf')
        model = SVC(kernel='rbf', C=1.0, gamma='scale')  # You can tune these later
        model.fit(X_train_sel, y_train)

        # Evaluate performance
        score_cv = model.score(X_cv_sel, y_cv)
        score_test = model.score(X_test_sel, y_test)

        # Print result
        print(f"\nFeatures: {feature_list}")
        print(f"  CV Score:   {score_cv}")
        print(f"  Test Score: {score_test}")
        print(f"  Model: SVM (kernel='rbf', C=1.0)")
```
